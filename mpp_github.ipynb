```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv("/content/passing-grade.csv", encoding='unicode_escape')

# Mengambil 30 baris pertama dari data
filtered_data = data.head(50)

# Menghitung jumlah kemunculan setiap nilai pada kolom 'PTN'
quantity = filtered_data['PTN'].value_counts()

# Warna yang akan digunakan pada setiap potongan diagram pie
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2', '#ffb366']

# Membuat diagram pie dengan ukuran yang lebih besar dan tambahan efek bayangan
plt.figure(figsize=(10, 10))
plt.pie(quantity, labels=quantity.index, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True, wedgeprops=dict(width=0.3))

# Menambahkan judul dan efek bayangan
plt.title('Top 30 Passing Grades by PTN', fontsize=16, fontweight='bold')
plt.gca().set_aspect('equal')  # Untuk memastikan diagram pie berbentuk lingkaran

# Menambahkan legenda dengan warna yang sesuai
legend_labels = [f'{label} ({count})' for label, count in zip(quantity.index, quantity)]

# Menampilkan diagram pie
plt.show()
```
